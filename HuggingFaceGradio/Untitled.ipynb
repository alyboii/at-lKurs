```python
import torch  # <--- WE ADDED THIS
import gradio as gr
from transformers import pipeline

# 1. Setup the model
# We explicitly tell it to use the model "distilbert" to stop the warning message
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
sentiment_model = pipeline("sentiment-analysis", model=model_name)

# 2. Define function
def analyze_text(text):
    result = sentiment_model(text)[0]
    return f"Label: {result['label']}, Score: {result['score']:.2f}"

# 3. Launch Interface
interface = gr.Interface(fn=analyze_text, inputs="text", outputs="text")
interface.launch()
```
