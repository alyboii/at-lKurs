#  Regression Analysis: The Ultimate Cheat Sheet

## 1. Types of Regression
Regression is the process of finding the relationship between Independent Variables ($X$) and a Dependent Target ($y$).

###  Simple Linear Regression
* **Definition:** Predicting using **only one** feature.
* **Formula:** $$y = b_0 + b_1x$$
* **Example:** Predicting *House Price* using only *Square Meters*.
* **Visual:** A single straight line on a 2D graph.

###  Multiple Linear Regression
* **Definition:** Predicting using **two or more** features.
* **Formula:** $$y = b_0 + b_1x_1 + b_2x_2 + ... + b_nx_n$$
* **Example:** Predicting *House Price* using *Square Meters*, *Room Count*, and *Location*.
* **Visual:** A flat plane (3D) or hyperplane (multi-dimensional).

---

## 2. Evaluation Metrics (The Report Card)
How do we know if our model is good? We check the **Accuracy** and the **Error**.

###  The "Success" Score: $R^2$ (R-Squared)
* **What it measures:** How well the model explains the variance in the data.
* **Range:** $0.0$ (Terrible) to $1.0$ (Perfect).
* **Formula:** $$R^2 = 1 - \frac{SS_{residual}}{SS_{total}}$$
* **Critical Note:** If you have many features, use **Adjusted $R^2$** to prevent overfitting.

###  The "Error" Scores (Lower is Better)

| Metric | Name | Formula | Why use it? |
| :--- | :--- | :--- | :--- |
| **MSE** | Mean Squared Error | $$\frac{1}{n} \sum (y_{true} - y_{pred})^2$$ | **Punishes big mistakes** heavily. Useful for training (Gradient Descent). |
| **RMSE** | Root Mean Squared Error | $$\sqrt{MSE}$$ | **Easier to read.** It converts the error back to the original unit (e.g., Dollars). |
| **MAE** | Mean Absolute Error | $$\frac{1}{n} \sum |y_{true} - y_{pred}|$$ | **Fair.** It ignores outliers and gives the average error. |

---

###  Pro Tip for Interviews
> **"Which metric should I choose?"**
> * If you need to explain the error to your manager in dollars/Euros, use **RMSE** or **MAE**.
> * If you want to force the model to fix large errors during training, use **MSE**.

#  Model Diagnosis: The Bias-Variance Tradeoff

In Machine Learning, we don't just want a model that works on the data we *have* (Training). We want a model that works on data we *don't have yet* (Testing).

## 1. The Three States of a Model

###  Scenario A: Underfitting (High Bias)
* **Analogy:** "The Lazy Student." The student didn't study at all.
* **Behavior:** The model is too simple to capture the pattern in the data.
* **Symptoms:**
    * Training Score: **Low** 
    * Test Score: **Low** 
* **Solution:** Add more features, make the model more complex.

###  Scenario B: Overfitting (High Variance)
* **Analogy:** "The Memorizer." The student memorized the textbook answers but fails the exam when questions change slightly.
* **Behavior:** The model captures the underlying pattern **AND** the random noise/errors in the data. It fits the training data perfectly but fails in the real world.
* **Symptoms:**
    * Training Score: **Very High** (e.g., 99%) 
    * Test Score: **Low** (e.g., 65%) 
* **Solution:** Get more data, remove complex features, or use Regularization.

###  Scenario C: Generalized (Balanced)
* **Analogy:** "The Smart Student." The student understands the *logic*, not just the answers.
* **Behavior:** The model ignores the noise and captures the true trend.
* **Symptoms:**
    * Training Score: **High** 
    * Test Score: **High** 

---

## 2. The Bias-Variance Tradeoff Table

We want to balance these two errors:

| Metric | Definition | High Bias Means... | High Variance Means... |
| :--- | :--- | :--- | :--- |
| **Bias** | How far the predictions are from the real values. | **Underfitting** (Too simple) | Good fitting |
| **Variance** | How much the prediction changes if we switch datasets. | Stable model | **Overfitting** (Unstable/Sensitive) |
