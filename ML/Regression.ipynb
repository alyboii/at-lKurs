```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# 1. Prepare the Data
# Imagine this data comes from a real Excel/CSV file
data = {
    'Size_m2': [50, 60, 80, 100, 120],       # Feature 1
    'Bedrooms': [1, 1, 2, 3, 3],             # Feature 2
    'Building_Age': [10, 5, 20, 10, 2],      # Feature 3
    'Price': [150, 180, 220, 300, 350]       # Target (Label)
}

df = pd.DataFrame(data)

# 2. Separate Features (X) and Target (y)
# X is now a matrix (table), not just a list
X = df[['Size_m2', 'Bedrooms', 'Building_Age']]
y = df['Price']

# 3. Initialize and Train
model = LinearRegression()
model.fit(X, y)

# 4. Make a Prediction
# House: 90 m2, 2 Bedrooms, 5 Years old
new_house = [[90, 2, 5]]
predicted_price = model.predict(new_house)

print(f"Predicted Price: ${predicted_price[0]:.2f}k")

# Let's see which factor is most important (Coefficients)
print("Coefficients (Importance):", model.coef_)
```
