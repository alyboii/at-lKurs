Interview Question 1: Data Cleaning (String to Number)

The Scenario: We want to find out which app is the most popular. To do this, we need to sort the data by the number of Installs.

However, if you look at the Installs column, the data looks like this: "10,000+", "500,000+", "Free". Pandas reads this as an Object (String), not a Number. We cannot do math on it.

My Question to You: How would you clean this column to make it an Integer? Which Python string functions would you use?

```python

df["Installs"]=df["Installs"].str.replace("+","",regex=False)
df["Installs"]=df["Installs"].str.replace(",","",regex=False)
df=df[df["Installs"]!="Free"]
```

```python
df["Installs"].unique()
```
